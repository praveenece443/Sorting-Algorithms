# QuickSort

<p>Runs O(n<sup>2</sup>) times</p>

<img src = attachment:quick%20sort.png width='456'></img>

```python
def quick_sort(A, left, right):
    if left < right:
        pivot_pos = partition(A, left, right)
        quick_sort(A, left, pivot_pos -1)
        quick_sort(A, pivot_pos +1, right)
    return A
        
def partition(A, left, right):
    i = left # i  starts from begining
    j = right -1 #j starts from last but -1 i.e., next to last
    pivot = A[right] #last value as pivot point
    while i < j:
        while i < right and A[i] < pivot:
            i += 1
        while j > left and A[j] >= pivot:
            j -= 1
        if i < j:
            A[i], A[j] = A[j], A[i]
    if A[i] > pivot:
        A[i], A[right] = A[right], A[i]
    return i
```

```python
%%time
A = [2,5,7,1,9,6,4]
quick_sort(A, 0, len(A)-1)
```
