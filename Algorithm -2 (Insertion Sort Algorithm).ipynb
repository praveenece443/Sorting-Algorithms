# Insertion Sort

<p>Runs O(n<sup>2</sup>) times</p>

```python
def insertion_sort(A):
    for i in range(1,len(A)):
        for j in range(i-1, -1, -1):
            if A[j] > A[j+1]:
                A[j], A[j+1] = A[j+1], A[j]
            else:
                break
    return A
```

```python
%%time
A = [2,5,7,1,9,6,4]
B = insertion_sort(A)
print(B)
```

### Using while loop

```python
def insertion_sort(A):
    for i in range(1,len(A)):
        j = i
        while A[j-1] > A[j] and j > 0:
            A[j-1], A[j] = A[j], A[j-1]
            j -= 1
    return A
```

```python
%%time
A = [2,5,7,1,9,6,4]
B = insertion_sort(A)
print(B)
```
