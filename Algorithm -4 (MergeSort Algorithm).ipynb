# Merge Sort (Divide and Sort Algorithm)

<p>Runs O(n*log(n)) times</p>
<p>It is a divide and conquer algorithm</p>

<div>

</div>

```python
def merge_sort(A):
    if len(A) > 1:
        left_arr = A[:len(A)//2]
        right_arr = A[len(A)//2:]
        merge_sort(left_arr)
        merge_sort(right_arr)
        
        i = 0
        j = 0
        k = 0
        while i < len(left_arr) and j < len(right_arr):
            if left_arr[i] < right_arr[j]:
                A[k] = left_arr[i]
                i += 1
            else:
                A[k] = right_arr[j]
                j += 1
            k += 1
        while i < len(left_arr):
            A[k] = left_arr[i]
            i += 1
            k += 1
        while j < len(right_arr):
            A[k] = right_arr[j]
            j += 1
            k += 1
    return A
```

```python
%%time
A = [2,6,4,5,1,3]
merge_sort(A)
```
